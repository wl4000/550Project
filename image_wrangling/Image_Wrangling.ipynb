```python
# List of the 32 countries in the 2018 World Cup
country_names = ['Brazil', 'Germany', 'Spain', 'Argentina', 'France',
                 'Portugal', 'England', 'Colombia', 'Belgium', 'Peru',
                 'Uruguay', 'Switzerland', 'Mexico', 'Croatia', 'Denmark',
                 'Poland', 'Sweden', 'Iran', 'Iceland', 'Serbia',
                 'Senegal', 'South Korea', 'Costa Rica', 'Australia', 'Morocco',
                 'Nigeria', 'Japan', 'Russia', 'Panama', 'Tunisia',
                 'Egypt', 'Saudi Arabia']
```

## Player Data

```python
import pandas as pd

# Load player data and only keep players from countries in the 2018 World Cup
raw_player_df = pd.read_csv('CompleteDataset.csv').drop('Unnamed: 0', axis=1).drop_duplicates()
raw_player_df = raw_player_df.dropna(subset=['ID', 'Name', 'Age', 'Nationality', 'Overall', 'Potential', 'Club', 'Value', 'Wage'])
raw_player_df['Nationality'] = raw_player_df['Nationality'].replace('Korea Republic', 'South Korea')
total_player_df = raw_player_df[raw_player_df['Nationality'].isin(country_names)].copy()
total_player_df
```

```python
# Country flag images
country_flags_df = total_player_df[['Nationality', 'Flag']].copy()
country_flags_df.columns = ['country', 'flag']
country_flags_df = country_flags_df.drop_duplicates()
country_flags_df = country_flags_df.reset_index()[['country', 'flag']]
country_flags_df
```

```python
flags_dict = {}
for i, row in country_flags_df.iterrows():
    flags_dict[row['country']] = row['flag']
flags_dict
```

```python
# Player photos
player_images_df = total_player_df[['ID', 'Name', 'Nationality', 'Photo']].copy()
player_images_df.columns = ['player_id', 'name', 'nationality', 'photo']
player_images_df = player_images_df.drop_duplicates().dropna()
player_images_df
```

```python
# Fix player photo links
player_images_df['photo'] = player_images_df['photo'].apply(lambda link : link.replace('org/48/', 'org/1x/'))
player_images_df    
```

```python
# Remove accents from player and club names
import unidecode

player_images_df['name'] = player_images_df['name'].apply(unidecode.unidecode)
player_images_df
```

```python
player_images_df.dtypes
```

```python
country_photo_dict = {}
for country in country_names:
    country_photo_dict[country] = []
for i, row in player_images_df.iterrows():
    country = row['nationality']
    player_id = row['player_id']
    name = row['name']
    photo = row['photo']
    country_photo_dict[country].append({'player_id': player_id, 'name': name, 'nationality': country, 'photo': photo})
# country_photo_dict
```

```python
photo_dict_list = []
for country in country_photo_dict:
    flag = flags_dict[country]
    players = country_photo_dict[country]
    photo_dict_list.append({'country': country, 'flag': flag, 'players': players})
# photo_dict_list
```

## Output to JSON

```python
# countryFlags
country_flags_df.to_json('countryFlags.json', orient='records')
```

```python
# playerPhotos
player_images_df.to_json('playerPhotos.json', orient='records')
```

```python
# countryPlayers
import json

with open('countryPlayers.json', 'w') as file:
    json.dump(photo_dict_list, file)
```
