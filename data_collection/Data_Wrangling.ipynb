```python
# List of the 32 countries in the 2018 World Cup
country_names = ['Brazil', 'Germany', 'Spain', 'Argentina', 'France',
                 'Portugal', 'England', 'Colombia', 'Belgium', 'Peru',
                 'Uruguay', 'Switzerland', 'Mexico', 'Croatia', 'Denmark',
                 'Poland', 'Sweden', 'Iran', 'Iceland', 'Serbia',
                 'Senegal', 'South Korea', 'Costa Rica', 'Australia', 'Morocco',
                 'Nigeria', 'Japan', 'Russia', 'Panama', 'Tunisia',
                 'Egypt', 'Saudi Arabia']
```

## Player Data

```python
import pandas as pd

# Load player data and only keep players from countries in the 2018 World Cup
raw_player_df = pd.read_csv('CompleteDataset.csv').drop('Unnamed: 0', axis=1).drop_duplicates()
raw_player_df = raw_player_df.dropna(subset=['ID', 'Name', 'Age', 'Nationality', 'Overall', 'Potential', 'Club', 'Value', 'Wage'])
raw_player_df['Nationality'] = raw_player_df['Nationality'].replace('Korea Republic', 'South Korea')
total_player_df = raw_player_df[raw_player_df['Nationality'].isin(country_names)].copy()
total_player_df
```

```python
# For each player, document what positions he plays
positions_dict_list = []
for i, row in total_player_df.iterrows():
    player_id = row['ID']
    for position in row['Preferred Positions'].strip().split(' '):
        overall = row['Overall'] if position == 'GK' else row[position]
        positions_dict_list.append({'player_id' : player_id, 'position' : position, 'overall' : overall})
```

```python
# Create DataFrame of players by their positions
player_position_df = pd.DataFrame(positions_dict_list)[['player_id', 'position', 'overall']]
player_position_df = player_position_df.dropna().drop_duplicates()
player_position_df['overall'] = player_position_df['overall'].astype(int)
player_position_df
```

```python
# Project chosen columns from player data to make a player DataFrame
player_df = total_player_df[['ID', 'Name', 'Age', 'Nationality', 'Overall', 'Potential', 'Club', 'Value', 'Wage']].copy()
player_df.columns = ['player_id', 'name', 'age', 'nationality', 'overall', 'potential', 'club', 'value', 'wage']
player_df = player_df.drop_duplicates().dropna()
player_df
```

```python
# Convert player values and wages to real integers
def convert_value(val):
    if val[-1] == 'M':
        return float(val[1:-1]) * 1000000
    elif val[-1] == 'K':
        return float(val[1:-1]) * 1000
    else:
        return 0
        
player_df['value'] = player_df['value'].apply(convert_value).astype(int)
player_df['wage'] = player_df['wage'].apply(convert_value).astype(int)
```

```python
player_df
```

```python
# Remove accents from player and club names
import unidecode

player_df['name'] = player_df['name'].apply(unidecode.unidecode)
player_df['club'] = player_df['club'].apply(unidecode.unidecode)
player_df
```

```python
player_df.dtypes
```

## Country Elo Data

```python
from bs4 import BeautifulSoup

# Open BeatifulSoup on html file containing country performance data
input_html = "World Football Elo Ratings.html"
with open(input_html, "r") as ifile:
    soup = BeautifulSoup(ifile, 'html5lib')
```

```python
# Find all relevant data in the html file
soup_result = soup.find_all('div', ['local-cell', 'global-cell', 'team-cell', 'rating-cell'])
soup_text = [tag.text for tag in soup_result]
soup_text = soup_text[soup_text.index('1') :]
```

```python
# Column names to be used for country DataFrame
country_ratings_schema = ['local_rank', 'global_rank', 'country', 'elo_rating', 'average_rank', 'average_elo_rating',
                          'rank_year_change', 'elo_rating_year_change', 'total_matches', 'home_matches',
                          'away_matches', 'neutral_matches', 'wins', 'losses', 'draws', 'goals_for', 'goals_against']

# Use the BeatifulSoup/html data to populate the country DataFrame
country_dict_list = []
for i, val in enumerate(soup_text):
    if i % len(country_ratings_schema) == 0:
        country_dict = {}
    country_dict[country_ratings_schema[i % len(country_ratings_schema)]] = val
    if i % len(country_ratings_schema) == len(country_ratings_schema) - 1:
        country_dict_list.append(country_dict)
country_df = pd.DataFrame(country_dict_list)[country_ratings_schema]
country_df
```

```python
# Convert weird negative sign to actual negative sign
def replace_negative(val):
    if val[0] == '−':
        return '-' + val[1:]
    else:
        return val
        
# Replace negative signs for yearly change of rank and elo rating
country_df['rank_year_change'] = country_df['rank_year_change'].apply(replace_negative)
country_df['elo_rating_year_change'] = country_df['elo_rating_year_change'].apply(replace_negative)

# Convert all columns except the country name to integers
for attr in country_ratings_schema:
    if attr != 'country':
        country_df[attr] = country_df[attr].astype(int)
```

```python
country_df.dtypes
```

```python
# Project chosen columns of country DataFrame
country_df = country_df[['local_rank', 'global_rank', 'country', 'elo_rating',
                         'rank_year_change', 'elo_rating_year_change', 'total_matches',
                         'wins', 'losses', 'draws', 'goals_for', 'goals_against']]
country_df
```

```python
# Store group information in a DataFrame
group_df = pd.read_csv('groups.csv')
group_df
```

```python
# Merge country_df and group_df to add group data for each country
country_df = country_df.merge(group_df, left_on='country', right_on='country')
country_df
```

## Country Economic Data

```python
# Parse economic data for each country and store in a list of dictionaries
with open('EconomicData.csv') as econ_file:
    lines = econ_file.read().splitlines()[1:]
    econ_dict_list = []
    for i in range(len(country_names)):
        country = lines[4 * i].split(',')[0]
        gdp = lines[4 * i].split(',')[-1]
        gdp_per_capita = lines[4 * i + 1].split(',')[-1]
        unemployment = lines[4 * i + 2].split(',')[-1]
        interest_rate = lines[4 * i + 3].split(',')[-1]
        econ_dict_list.append({'country' : country, 'gdp' : gdp, 'gdp_per_capita' : gdp_per_capita,
                               'unemployment_rate' : unemployment, 'interest_rate' : interest_rate})
# Store econ data in a DataFrame
econ_df = pd.DataFrame(econ_dict_list)
econ_df
```

```python
# Fix country names in econ_df
econ_df['country'] = econ_df['country'].replace(
    {'United Kingdom' : 'England', '"Iran' : 'Iran', '"Korea' : 'South Korea',
     'Russian Federation' : 'Russia', '"Egypt' : 'Egypt'})
```

```python
econ_df.dtypes
```

```python
# Replace missing values with NaNs
def nullify(val):
    if val == '..':
        return float('NaN')
    else:
        return val
    
econ_df['unemployment_rate'] = econ_df['unemployment_rate'].apply(nullify)
```

```python
# Convert economic data to float type
econ_df['gdp'] = econ_df['gdp'].astype(float)
econ_df['gdp_per_capita'] = econ_df['gdp_per_capita'].astype(float)
econ_df['interest_rate'] = econ_df['interest_rate'].astype(float)
econ_df['unemployment_rate'] = econ_df['unemployment_rate'].astype(float)
```

```python
econ_df
```

```python
econ_df.dtypes
```

```python
# Merge country elo data and economic data into one DataFrame
country_full_df = country_df.merge(econ_df, left_on='country', right_on='country')
country_full_df
```

## World Cup Participation Data

```python
# Load data of past soccer matches into a DataFrame
match_results_df = pd.read_csv('results.csv')
match_results_df['home_team'] = match_results_df['home_team'].replace('Korea Republic', 'South Korea')
match_results_df['away_team'] = match_results_df['away_team'].replace('Korea Republic', 'South Korea')
match_results_df
```

```python
# Limit to world cup matches involving one of the teams in the 2018 World Cup
world_cup_matches_df = match_results_df[match_results_df['tournament'] == 'FIFA World Cup']
matches_df = world_cup_matches_df[world_cup_matches_df['home_team'].isin(country_names)
                                      | world_cup_matches_df['away_team'].isin(country_names)].copy()
matches_df
```

```python
# Convert the date of each match to the corresponding year
matches_df['date'] = matches_df['date'].apply(lambda date : date.split('-')[0])
matches_df = matches_df.rename(columns={'date' : 'year'})
matches_df
```

```python
# Use the home_team column to find the years in which each country participated in a World Cup
home_team_df = matches_df.merge(country_df, left_on='home_team', right_on='country')[['year', 'home_team']]
home_team_df = home_team_df[home_team_df['home_team'].isin(country_names)]
home_team_df = home_team_df.drop_duplicates().rename(columns={'home_team' : 'country'})
home_team_df
```

```python
# Use the away_team column to find the years in which each country participated in a World Cup
away_team_df = matches_df.merge(country_df, left_on='away_team', right_on='country')[['year', 'away_team']]
away_team_df = away_team_df[away_team_df['away_team'].isin(country_names)]
away_team_df = away_team_df.drop_duplicates().rename(columns={'away_team' : 'country'})
away_team_df
```

```python
# Concatenate home_team_df and away_team_df to create complete participated_in data
participated_in_df = pd.concat([home_team_df, away_team_df]).drop_duplicates().sort_values('year')
participated_in_df['year'] = participated_in_df['year'].astype(int)
participated_in_df
```

```python
# Check how many World Cups each country participated in (used to verify accuracy of participated_in_df)
participated_in_df.groupby('country').size().reset_index(name='count').sort_values('count', ascending=False)
```
